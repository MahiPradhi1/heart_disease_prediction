```python
df['Heart Disease Status'] = df['Heart Disease Status'].map({ 'Yes': 1,  'No': 0})
```

```python
num_cols = [
    'Age', 'Blood Pressure', 'Cholesterol Level', 'BMI',
    'Sleep Hours', 'Triglyceride Level', 'Fasting Blood Sugar',
    'CRP Level', 'Homocysteine Level'
]

for col in num_cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
cat_cols = [
    'Gender', 'Exercise Habits', 'Smoking', 'Family Heart Disease',
    'Diabetes', 'High Blood Pressure', 'Low HDL Cholesterol',
    'High LDL Cholesterol', 'Alcohol Consumption', 'Stress Level',
    'Sugar Consumption'
]

for col in cat_cols:
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
# Total number of heart disease patients
total_hd = df[df['Heart Disease Status']==1].shape[0]

# List of bad habits
bad_habits = ['Low Exercise', 'Short Sleep', 'High Alcohol', 'High Sugar']

# Calculate percentage for each bad habit
percent_hd_bad_habits = [
    (df[(df['Exercise Habits'] == 'Low') & (df['Heart Disease Status']==1)].shape[0] / total_hd) * 100,
    (df[(df['Sleep Hours'] == '<6') & (df['Heart Disease Status']==1)].shape[0] / total_hd) * 100,
    (df[(df['Alcohol Consumption'] == 'High') & (df['Heart Disease Status']==1)].shape[0] / total_hd) * 100,
    (df[(df['Sugar Consumption'] == 'High') & (df['Heart Disease Status']==1)].shape[0] / total_hd) * 100
]

print(percent_hd_bad_habits)
```

```python
plt.pie(percent_hd_bad_habits, labels=bad_habits)
plt.show()
```
